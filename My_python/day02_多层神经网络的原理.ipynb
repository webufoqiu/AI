# $$sigmoid= \sigma(x)=\frac{1}{1+e^(-x)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))
sub_x= np.linspace(-10,10)
plt.plot(sub_x,sigmoid(sub_x))
```

```python

import random 
def random_linear(x):
    k,b=random.random(),random.random()
    print(k,b)
    return k*x+b

def sigmoid(x):
    return 1/(1+np.exp(-x))

sub_x2= np.linspace(-10,10)

for _  in range(5):
    plt.plot(sub_x2,random_linear(sigmoid(sub_x2)))
```
